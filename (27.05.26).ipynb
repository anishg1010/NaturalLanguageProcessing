```python
import gensim.downloader as api
```

```python
model = api.load("word2vec-google-news-300")
```

```python
import tensorflow as tf
```

```python
# Example usage
word = "king"       
embedding_vector = model[word]
print(f"Embedding vector for '{word}': {embedding_vector}") 
# Convert the embedding vector to a TensorFlow tensor
embedding_tensor = tf.convert_to_tensor(embedding_vector, dtype=tf.float32)
print(f"TensorFlow tensor for '{word}': {embedding_tensor}")
# Example usage in a TensorFlow model
# Assuming you have a simple model that takes the embedding as input
input_layer = tf.keras.layers.Input(shape=(300,))  # 300-dimensional embedding
dense_layer = tf.keras.layers.Dense(64, activation='relu')(input_layer)
model = tf.keras.Model(inputs=input_layer, outputs=dense_layer)
# Now you can use the model with the embedding tensor
output = model(embedding_tensor[None, :])  # Add batch dimension
print(f"Output from the model: {output}")
```
