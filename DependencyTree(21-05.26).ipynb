```python
text = "this is a complex sentence with a lot of words and a lot of dependencies. it is a very long sentence and it is a very complex sentence. Let's see how it works."
```

```python
from scipy.spatial import distance
string_1 = "Table"
string_2 = "Taken"

hamming_distance = distance.hamming(list(string_1), list(string_2))
print(f"Hamming Distance: {hamming_distance}")  


Normalized_hamming_distance = hamming_distance / max(len(string_1), len(string_2))
print(f"Normalized Hamming Distance: {Normalized_hamming_distance}")
```

```python
doc1 = "This is a sentence about machine learning."
doc2 = "This sentence is about deep learning."

words1 = doc1.lower().split()
words2 = doc2.lower().split()    

print(f"Words in Document 1: {words1}")
print(f"Words in Document 2: {words2}") 
set_doc1 = set(doc1.split())    
set_doc2 = set(doc2.split())
print(f"Unique Words in Document 1: {set_doc1}")
print(f"Unique Words in Document 2: {set_doc2}")

intersection = set_doc1.intersection(set_doc2)
union = set_doc1.union(set_doc2)    
jaccard_similarity = len(intersection) / len(union)
print(f"Jaccard Similarity: {jaccard_similarity:.4f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = ["I love machine learning.",
             "I love natural language processing.",
             "Machine learning is great.",]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)    \

print("VOcabulary:")
print(vectorizer.get_feature_names_out())  
print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(tfidf_matrix) 
print("Cosine Similarity Matrix:")
print(similarity_matrix)

similarity_score = similarity_matrix[0][1]
print(f"Cosine Similarity between Document 1 and Document 2: {similarity_score:.4f}")
```
