```python
import spacy
from spacy import displacy
from IPython.display import HTML, display

#   Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

# Print the dependency parse information for each token
for token in doc:
    print(f"{token.text:<12} {token.dep_:<12} {token.head.text}")

html = displacy.render(doc, style="dep", jupyter=False)

# Display the dependency parse visualization in Jupyter Notebook
display(HTML(html))

# Discuss with your partner:
# 1. What is the ROOT? 
# saw

# 2. What does "with a telescope" modify? 
# the mans possesion or the mans description

# 3. Does the sentence have more than one meaning? 
# yes

# 4. Did spaCy choose the interpretation you expected? 
# yes

# 5. Rewrite the sentence to remove ambiguity. 
# I saw the man using a telescope.
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative
in Berlin on 15 June 2026.

The project involves collaboration with Google,
SAP and SRH University Heidelberg.

Satya Nadella will present the initiative
at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text,
        ent.label_,
        spacy.explain(ent.label_)
    )

#1. Which entities were detected? 
# Microsoft, Berlin, 15 June 2026, Google, SAP, SRH University Heidelberg, Satya Nadella, Germany

#2. Which labels were assigned?
# ORG, GPE, DATE, ORG, ORG, ORG, PERSON, GPE

#3. Which entity types occur most often?
# ORG, GPE

#4. Were any entities missed?
# No, all entities were detected.
     
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)
print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))

# 1. Which pair has the highest similarity?
# Doc1 vs Doc2 has the highest similarity (0.707) because they both discuss healthcare and AI/machine learning, while Doc3 is about football and has low similarity with the other two documents.

# 2. Is this result reasonable?
# Yes, it is reasonable because Doc1 and Doc2 share common themes related to healthcare and technology, while Doc3 is on a completely different topic (sports), leading to low similarity scores.

# 3. Why does TF-IDF treat some words as more important?
# TF-IDF assigns higher importance to words that are unique to a document and appear less frequently across the entire corpus. 
# Common words that appear in many documents (like "is", "a", "the") are given lower weights, while more specific terms (like "AI", "healthcare") that are less common across documents receive higher weights, indicating their importance in distinguishing the content of the document.

# 4. Does cosine understand meaning?
# No, cosine similarity does not understand meaning. It only measures the similarity of the vector representations of the documents based on the presence and frequency of words. It does not capture the semantic relationships or context of the words, so it cannot understand the meaning behind the text.

# 5. Would it know that "AI" and "machine learning" are related?
# No, cosine similarity would not inherently know that "AI" and "machine learning" are related. It would only consider their co-occurrence in the documents and their relative frequencies. 
```
